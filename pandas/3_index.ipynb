# Pandas objects index
By default index in pandas objects is a range from 0 to $n - 1$ where $n$ is a number of observations  
Here we will make it more meaningful

```python
import numpy as np
import pandas as pd
# Restricting number of displaying rows, just for convenience
pd.set_option('max_rows', 8)
```

```python
default_index = pd.read_csv('data/movie.csv')
default_index.head()
```

One way to set index in your dataframe is to specify `index_col` argument

```python
movie = pd.read_csv('data/movie.csv', index_col='movie_title')
movie.head()
```

Other way is to set index after reading dataset with method `set_index`

```python
default_index = default_index.set_index('movie_title')
```

`set_index` has following arguments  
`set_index(keys, drop=True, append=False, inplace=False, verify_integrity=False)` and by default returns new dataframe. You can make inplace change by specifying inplace argument equal to `True`  
`drop` - whether to remove column which is used as new index from DataFrame  
`append` - whether to add new columns to already existing index  
`verify_integrity` - check index values uniqueness (reduce performance)

To replace index with default numeric index `reset_index` method can be used.  
Note by default it places old index column to the 1st position in DataFrame

```python
default_index.reset_index(inplace=True)
default_index.head()
```

## Columns and index renaming
Coolumns and index can be obtained via attributes

```python
movie.index
movie.columns
```

Each of them return a modified array containing index and column names.

To rename them you could use cool method `rename` which takes dictionary with 1-to-1 mapping of old to new names.  
Alternatively it can apply function to names.

```python
new_cols = {'num_user_for_reviews': 'user_reviews', 
           'num_critic_for_reviews': 'critic_reviews'}
movie.rename(columns=new_cols, inplace=True)
movie.head()
```

```python
# Write all column names in upper case
# With axis=0 it would rewrite index labels
movie.rename(str.upper, axis=1)
```
